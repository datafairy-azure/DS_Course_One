```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
# Functional Matplotlib
plt.plot(x,y,'r-')
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
# OO Matplotlib
fig = plt.figure()

axes = fig.add_axes([0.1,0.1,0.8,0.8])

axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])

axes1.plot(x,y)
axes2.plot(y,x)
axes1.set_title('Quad')
axes2.set_title('Polynomial')
```

```python
fig,axes = plt.subplots(nrows=1,ncols=2)

axes[0].plot(x,y)
axes[0].set_title('Quad')
axes[1].plot(y,x)
axes[1].set_title('Poly')

plt.tight_layout()
```

```python
fig = plt.figure(figsize=(8,2),dpi=100)
# dpi = dots per inch
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,ax = plt.subplots(nrows=2,ncols=1,figsize=(8,2),dpi=100)
# dpi = dots per inch
ax[0].plot(x,y)
ax[1].plot(y,x)

plt.tight_layout()
```

```python
fig.savefig('my_picture.png')
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X^2',color='#FF8C00',linewidth=1.5,ls='--')
ax.plot(x,x**3, label='X^3',color='purple',lw=0.5, alpha=0.3,ls='steps')
ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x**2,label='X^2',color='#FF8C00',marker='o',markerfacecolor='yellow', markeredgewidth=1, markeredgecolor='b')
ax.plot(x,x**3, label='X^3',color='purple',marker='x',markersize=10)
ax.legend()
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**3, label='X^3',color='purple',marker='x',markersize=10)
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
